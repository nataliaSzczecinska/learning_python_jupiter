Stwórz dane z rozkładu normalnego o średniej 3 i odchyleniu standardowym 1 dla 100 oraz 10 000. Na podstawie tych sztucznie stworzonych zbiorów oblicz statystyki poznane w module statystki opisowej. Opisz różnice pomiędzy statystkami i wynik swojej pracy wyślij Mentorowi jako Jupyter Noteboook zapisany w Github.

```python
# manipulacja danymi
import numpy as np
import pandas as pd

# wizualizacja danych
import matplotlib.pyplot as plt
import seaborn as sns

# statystyczna analiza danych
from scipy import stats
```

```python
def gauss(x, sigma = 1, mu = 0):
    return stats.norm.pdf(x, mu, sigma)
```

```python
avg_value = 3
std_value = 1
x_100 = np.linspace(-10 + 3, 10 + 3, 100)
x_10000 = np.linspace(-10 + 3, 10 + 3, 10000)
y_100 = gauss(x_100, sigma = std_value, mu = avg_value)
y_10000 = gauss(x_10000, sigma = std_value, mu = avg_value)
plt.figure(figsize=(12, 8))
plt.plot(x_100, y_100, label='100 pkt, odchylenie standardowe = 1, średnia = 3')
plt.plot(x_10000, y_10000, label='10000 pkt, odchylenie standardowe = 1, średnia = 3')
plt.legend(loc='upper left')
plt.show()
```

```python
# data analysis
mean_100 = np.mean(y_100)
median_100 = np.median(y_100)
mode_100 = stats.mode(y_100)[0][0]
min_100 = y_100.min()
max_100 = y_100.max()
q0_100 = np.quantile(y_100, 0.0)
q1_100 = np.round(np.quantile(y_100, 0.25))
q2_100 = np.round(np.quantile(y_100, 0.5))
q3_100 = np.round(np.quantile(y_100, 0.75))
q4_100 = np.round(np.quantile(y_100, 1.0))
standard_deviation_100 = np.std(y_100, ddof=1)

mean_10000 = np.mean(y_10000)
median_10000 = np.median(y_10000)
mode_10000 = stats.mode(y_10000)[0][0]
min_10000 = y_10000.min()
max_10000 = y_10000.max()
q0_10000 = np.quantile(y_10000, 0.0) # min alternative
q1_10000 = np.round(np.quantile(y_10000, 0.25))
q2_10000 = np.round(np.quantile(y_10000, 0.5))
q3_10000 = np.round(np.quantile(y_10000, 0.75))
q4_10000 = np.round(np.quantile(y_10000, 1.0)) # max alternative
standard_deviation_10000 = np.std(y_10000, ddof=1)

print("STATISTIC ANALYSIS FOR 100 POINTS IN GAUSS")
print(f"Mean = {mean_100}, median = {median_100}, moda = {mode_100}")
print(f"min = {min_100}, max = {max_100}, standard deviation = {standard_deviation_100}")
print(f"Q0: {q0_100}, Q1: {q1_100}, Q2: {q2_100}, Q3: {q3_100}, Q4: {q4_100}")
print("STATISTIC ANALYSIS FOR 10000 POINTS IN GAUSS")
print(f"Mean = {mean_10000}, median = {median_10000}, moda = {mode_10000}")
print(f"min = {min_10000}, max = {max_10000}, standard deviation = {standard_deviation_10000}")
print(f"Q0: {q0_10000}, Q1: {q1_10000}, Q2: {q2_10000}, Q3: {q3_10000}, Q4: {q4_10000}")
```

```python
# prepare data to analysis
labels  = ['mean', 'median', 'mode', 'min', 'max', 'std', 'q1', 'q3']
data_100 = [mean_100, median_100, mode_100, min_100, max_100, standard_deviation_100, q1_100, q3_100]
data_10000 = [mean_10000, median_10000, mode_10000, min_10000, max_10000, standard_deviation_10000, q1_10000, q3_10000]
serie_100 = pd.Series(data_100, labels)
serie_10000 = pd.Series(data_10000, labels)
data = pd.DataFrame({'data_100' : data_100, 'data_10000' : data_10000}, index = labels)
data
```

```python
# bar plot
width = 0.4
index = np.arange(len(labels))
plt.bar(index - width/2, data_100, width, label = '100 pkt')
plt.bar(index + width/2, data_10000, width, label = '10000 pkt')
plt.xticks(index, labels)
plt.legend()
plt.show()
```

```python
def difference_between_100_and_1000(row):
    return abs(row['data_100'] - row['data_10000'])
   
data['difference'] = data.apply(lambda row: difference_between_100_and_1000(row), axis = 1)
max_from_difference = data['difference'].max()
max_from_difference_idx = data['difference'].idxmax()
print(f"Max difference = {max_from_difference} for {max_from_difference_idx}")
data
```

Różnice są w wartości średniej, mediany, wartości maksumalnej i odchylenia standardowego. Największa różnica jest w wartości maksymalnej, która wynosi 0.002.
