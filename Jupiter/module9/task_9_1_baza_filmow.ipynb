### 10 best movies by vote average, where vote count is more then 3rd quantile

```python
movies.describe()
```

```python
vote_count_3rd_quantile = movies['vote_count'].quantile(0.75)
print(f"The 3rd quantile of vote_count: {vote_count_3rd_quantile}")
```

```python
list_of_10_best_movies = movies[movies['vote_count'] > vote_count_3rd_quantile].sort_values(by = ['vote_average'], ascending = False)[:10]
list_of_10_best_movies
```

### Group movies for films between 2010 and 2016 to find avg revenue and avg budget

```python
# change 'release_date' type to datatime
movies['release_date'] = pd.to_datetime(movies['release_date'])
# group by year for avg revenue and avg budget
movies_group_by_avg_revenue_budget = movies.groupby(movies['release_date'].dt.year).agg({'budget':'mean','revenue':'mean'})
# take movies realised between 2010 and 2016
movies_group_by_avg_revenue_budget = movies_group_by_avg_revenue_budget[(movies_group_by_avg_revenue_budget.index <= 2016) & (movies_group_by_avg_revenue_budget.index >= 2010)]
movies_group_by_avg_revenue_budget
```

```python
# create a plot

# data
release_date = movies_group_by_avg_revenue_budget.index.to_series()
revenue = movies_group_by_avg_revenue_budget['revenue']
budget = movies_group_by_avg_revenue_budget['budget']

# function to recalculate value
def million(x, pos):
        return '$ {:2.1f}M'.format(x*1e-6)

#def plot
fig, ax = plt.subplots()
formatter = plt.FuncFormatter(million)
ax.yaxis.set_major_formatter(formatter)
ax.bar(release_date, revenue, label='revenue')
ax.plot(release_date, budget, label='budget', color = 'red')
ax.set_title('Średni przychód i budżet filmu w latach 2010 - 2016')
ax.legend(loc = (1.05, 0.0))
plt.show()
```

### Analysis of movie genres

```python
# join movies and genres
genres = genres.rename(columns = {'Unnamed: 0' :'genre_id'})
genres.head()
```

```python
movies.head()
```

```python
merge_movies_genres = pd.merge(movies, genres, how='inner',on='genre_id')
merge_movies_genres.head()
```

```python
group_by_genres = merge_movies_genres.groupby('genres').agg({'id':'count','runtime':'mean'})
group_by_genres = group_by_genres.rename(columns = {'id' : 'number_of_movies'})
group_by_genres
```

```python
genres_max = group_by_genres['number_of_movies'].idxmax()
genres_max_movies_count = group_by_genres['number_of_movies'].max()
print (f"Genres \'{genres_max}\' is the most popular. There are {genres_max_movies_count} movies of this type in data")
```

```python
genres_runtime_avg_max = group_by_genres['runtime'].idxmax()
genres_runtime_avg_for_max = group_by_genres['runtime'].max()
print (f"The longest average film duration was for the \'{genres_runtime_avg_max}\' genre. It is {genres_runtime_avg_for_max} minutes (?)")
```

```python
historical_movies = merge_movies_genres[merge_movies_genres['genres'] == genres_runtime_avg_max]
historical_movies.head()
```

```python
historical_movies['runtime'].hist(bins = 10, grid = False)
```
