```python
# import package
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
init_notebook_mode(connected = True)
import pandas as pd
import matplotlib.pyplot as plt
import plotly.offline as pyo
import plotly.graph_objs as go
from plotly import subplots
import plotly.figure_factory as ff
```

```python
# import data
kghm = pd.read_csv('kgh_d.csv')
cu_price = pd.read_csv('ca_c_f_d.csv')
display(kghm.head())
display(cu_price.head())
```

```python
# get information about data frames
display(kghm.info())
display(cu_price.info())
```

```python
# prepared data to visualisation
kghm['Data'] = pd.to_datetime(kghm['Data'])
display(kghm.head())
cu_price['Data'] = pd.to_datetime(cu_price['Data'])
display(cu_price.head())
display(kghm.info())
display(cu_price.info())
```

```python
data = pd.merge(kghm[['Data', 'Zamkniecie']], cu_price[['Data', 'Zamkniecie']], how = 'outer', on = 'Data')
data.rename(columns = {'Zamkniecie_x' : 'KGHM', 'Zamkniecie_y' : 'Cu'}, inplace = True)
data.head()
```

```python
display(data.info())
```

```python
# prepared data to plot
x_kghm = kghm['Data']
y_kghm = kghm['Zamkniecie']
x_cu = cu_price['Data']
y_cu = cu_price['Zamkniecie']
```

```python
#create plot
fig = go.Figure()
fig = subplots.make_subplots(rows = 2, cols = 1, subplot_titles=['KGHM','Cu'], shared_xaxes = True)
fig.add_trace(go.Scatter(x = x_kghm, y = y_kghm), 1, 1)
fig.add_trace(go.Scatter(x = x_cu, y = y_cu), 2, 1)
fig.update_layout(title='Zestawienie danych KGHM i Cu', showlegend = False)
iplot(fig)
```

```python
# table using pandas
data = data.sort_values(by = 'Data').reset_index()
data.drop(columns = 'index', inplace = True)
data
```

```python
# table using plotly
fig =  ff.create_table(data)
fig.show()
```
